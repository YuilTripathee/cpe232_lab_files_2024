*(Hint: The Fibonacci sequence is the series of numbers where each number is the sum of the two preceding numbers. For example, 0, 1, 1, 2, 3, 5, ...)*

```python
# write your code here
# standard function
def fibo(n):
    fibo_series = [0, 1]
    while len(fibo_series) < n:
        fibo_series.append(fibo_series[-1] + fibo_series[-2])
    return fibo_series

fibo_series = pd.Series(fibo(50))
fibo_series
```

### Combine Dataframe by .merge and .concat

3.6 Merge

```python
superstore_order = pd.read_csv('superstore_order.csv')
superstore_people = pd.read_csv('superstore_people.csv')
superstore_return = pd.read_csv('superstore_return.csv')
```

I would like to check the structure of each dataframe as we import the CSV files.

```python
superstore_order.head(1)
```

```python
superstore_people.describe()
```

```python
# All the value in Returned has Yes
superstore_return.loc[superstore_return.Returned != 'Yes']
```

```python
superstore_order.merge(superstore_return[superstore_return["Returned"]=="Yes"],
 on="Order ID" ,
 how="inner")\
 [["Customer ID", "Returned"]]\
 .drop_duplicates()
```

**[Question]** In your opinion, what information that the result above conveys?

Ans: This concept of merging is quite similar to the way we join different relations to form a view in relational database systems in my experience. We used inner join which means Order ID should match in both the relations. If Order ID present in one table (dataframe here) is not consistent to the other, it won't merge and just discharge.

We are merging with the condition if the ordered "thing" has "Yes" as its value in `superstore_return` dataframe. Just FYI, all the rows in that dataframe has `Yes` value for `Returned` column.

More merging...

```python
superstore_order.merge(superstore_return,
 on="Order ID" ,
 how="inner")
```

### Groupby

```python
# superstore_order.groupby(['Segment','Ship Mode', 'Category'])[['Sales','Quantity','Discount','Profit']].sum()
superstore_order.groupby(['Segment','Ship Mode'])[['Sales','Quantity','Discount','Profit']].sum()
```

**[Question]** Briefly describe an information that the result above conveys?

Ans: First, we group by segment followed by ship mode. Then, to clarify I also added category to the grouping so as to rearrange those rows that matches segment and the ship mode. Finally we added all the sales, discount and profit applied. This allows us to perform statistical analysis across the rows.

```python
superstore_order["Profit Ratio"] = superstore_order["Profit"]/superstore_order["Sales"]
```

```python
superstore_order.groupby(["Category", "Sub-Category"]).agg(mean_profit_ratio = ("Profit Ratio", "mean"))
```

**[Question]** Briefly describe an information that the result above conveys?

Ans: First, we have profit ratio being calculated by profit divided by sales. Then we aggregate the mean product ratio in each group that we crated using `.groupby()` function. Mean profit ratio of each group (`category --> sub-category`) is shown adjacently.

### Pivot and Melt

Pivot

```python
# try any number of states (FYI: there are 51 states in United States)
# put 52 and you'll still get 51 states (all of them)
superstore_order.pivot_table(index="State", columns="Ship Mode", values="Order ID", aggfunc="count").fillna(0).head(5)

# we're just counting orders in each state (indexed by state)
```

```python
pivot_table_result = superstore_order.pivot_table(index="State", columns="Ship Mode", values="Order ID", aggfunc="count").fillna(0)
print(pivot_table_result)
```

Melt

```python
melted_result = pd.melt(pivot_table_result.reset_index(), id_vars=["State"], var_name="Ship Mode", value_name="Order Count")
print(melted_result)

# massage the dataframe (according to docs 😄)
# returns unpivoted dataframe (that's all I want to know)
```
